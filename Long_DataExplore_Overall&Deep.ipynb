### Check the correlation and the covariance in the dataset, choosing the most suitable graph type to visualize the corr and cov matrix.

```python
corr = df1.corr()
corr
```

```python
cov = df1.cov()
cov
```

```python
# Using heatmap to visualize the corr and cov matrix
import seaborn as sns
plt.figure(figsize = (14,8))
plt.title("Correlation matrix of wine features")
sns.heatmap(corr, annot = True)
```

```python
plt.figure(figsize = (14,8))
plt.title("Covariation matrix of wine features")
sns.heatmap(cov, annot = True)
```

## Deep Analysis

### Using the histogram to visualize all feature in the dataset in form grid map 2x6

```python
features = list(df1.columns)
features
```

```python
df1[features].plot(kind='hist', subplots=True, layout=(3,4), figsize=(16,8))
```

```python
# Another way
plt.figure(figsize=(16,8))
for i in range(12):
    plt.subplot(2, 6, i+1)
    plt.title(features[i])
    plt.ylabel('Frequency')
    df1[features[i]].hist()
```

### Using scatter graph to visualize the relationship between two arbitrary features

```python
plt.title('Scatter graph of the relationship between two arbitrary features')
for i in range(12):
    for j in range(i+1,12):
        df1.plot(kind='scatter', x=features[i], y=features[j],  layout=(11,6), figsize=(10,6))
```
